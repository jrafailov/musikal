# Model Accuracy Comparison

Horizontal bar chart of pair-classification test accuracy across all model variants we tried. Tweak `font_scale` if labels need to be even bigger for the poster.

```python
import matplotlib.pyplot as plt
import numpy as np

models = [
    "PCA latent + RF",
    "RF default (100 trees)",
    "RF tuned (500 trees, depth 20)",
    "Gradient Boosting (best config)",
    "Siamese (early stopping)",
]
accuracies = [70.5, 72.6, 74.4, 77.0, 78.0]

colors = plt.cm.viridis(np.linspace(0.75, 0.35, len(models)))
```

```python
plt.rcParams["font.family"] = ["Helvetica", "Nimbus Sans", "Arial", "sans-serif"]

font_scale = 1.6

fig, ax = plt.subplots(figsize=(9, 5.5))
bars = ax.barh(models, accuracies, color=colors, edgecolor="none")

for bar, acc in zip(bars, accuracies):
    ax.text(
        bar.get_width() + 0.6,
        bar.get_y() + bar.get_height() / 2,
        f"{acc:.1f}%",
        va="center",
        fontsize=14 * font_scale,
    )

ax.set_xlim(0, 100)
ax.set_xlabel("Test Accuracy", fontsize=16 * font_scale)
ax.set_ylabel("Model", fontsize=16 * font_scale)
ax.tick_params(axis="y", labelsize=12 * font_scale)
ax.tick_params(axis="x", labelsize=13 * font_scale)
ax.grid(axis="x", linestyle="--", alpha=0.5)
ax.set_axisbelow(True)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.savefig("model_accuracy_comparison.png", dpi=200, bbox_inches="tight")
plt.show()
```
