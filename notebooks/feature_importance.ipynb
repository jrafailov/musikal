# Top 10 Feature Importances

Horizontal bar chart of the top 10 feature importances from the tuned RF baseline. Same styling as `model_accuracy_comparison.ipynb` so the poster reads as one set.

```python
import matplotlib.pyplot as plt
import numpy as np

features = [
    "Tempo",
    "Tempogram 1",
    "MFCC 0",
    "Tempogram 6",
    "Tempogram 8",
    "Tempogram 4",
    "MFCC 6",
    "Tonnetz Std 1",
    "RMS Mean",
    "Tempogram 5",
]
importances = [33.1, 4.6, 3.4, 3.2, 3.0, 2.7, 2.4, 2.3, 1.7, 1.7]

colors = plt.cm.viridis(np.linspace(0.75, 0.35, len(features)))
```

```python
plt.rcParams["font.family"] = ["Helvetica", "Nimbus Sans", "Arial", "sans-serif"]

font_scale = 1.6

fig, ax = plt.subplots(figsize=(9, 5.5))
bars = ax.barh(features, importances, color=colors, edgecolor="none")
ax.invert_yaxis()

for bar, val in zip(bars, importances):
    ax.text(
        bar.get_width() + 0.6,
        bar.get_y() + bar.get_height() / 2,
        f"{val:.1f}%",
        va="center",
        fontsize=14 * font_scale,
    )

ax.set_xlim(0, 50)
ax.set_xlabel("Importance", fontsize=16 * font_scale)
ax.set_ylabel("Feature", fontsize=16 * font_scale)
ax.tick_params(axis="y", labelsize=12 * font_scale)
ax.tick_params(axis="x", labelsize=13 * font_scale)
ax.grid(axis="x", linestyle="--", alpha=0.5)
ax.set_axisbelow(True)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.savefig("feature_importance.png", dpi=200, bbox_inches="tight")
plt.show()
```
